### Filter and Map function

1.filter() creates a new iterable by applying a filtering function to each element of the input iterable, returning only the elements that satisfy a given condition.

2.map() applies a transformation function to each element of the input iterable, creating a new iterable with the transformed values.

```python
df= pd.DataFrame({'Name': ['albin', 'harry', 'Chipmunks',"potter"], 'Age': [28, 10, 23,32]})
df
```

```python
filtered_df = df.filter( 'Age <= 23')
filtered_df
```

```python
lst= [1,2,23,24,3,34,37,39,58,59,60]
```

```python
result = list(filter(lambda x: x %2  == 0, lst))
```

```python
print(result)
```

### map()

```python
#It Returns a list of the results after applying the given function to each item of a given iterable (list, tuple etc.)
```

```python
incremented_value= list(map(lambda x: x+2, lst))
```

```python
incremented_value
```

### Merge function()

```python
import pandas as pd
```

```python
df1 = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})
df2 = pd.DataFrame({'A': [1, 6], 'D': [7, 8]})
```

```python
df = pd.merge(df1, df2, on='A', how='inner')
```
